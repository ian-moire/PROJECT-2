# Analysis of Healthcare Facility in Kenya

## Background
The healthcare sector in Kenya is rapidly evolving, with an increasing demand for quality healthcare services and improved health outcomes. The management and analysis of healthcare data are essential to enhance service delivery, optimize resource allocation, and ensure equitable access to healthcare services across the country. This project aims to provide insights into the distribution and characteristics of healthcare facilities in Kenya.

## Objective
The objective of this project is to analyze the healthcare facilities in Kenya, focusing on their geographic distribution, ownership, operational status, and key metrics such as the number of beds and cots. The analysis aims to identify trends and patterns for improvement in the healthcare sector.

## Data Description
The dataset used for this project, ehealth_facilities_repaired.xls, which was found on the openAFRICA website, contains data on 10,505 healthcare facilities in Kenya. The dataset includes various attributes such as facility code, name, province, county, district, type, owner, location details, number of beds and cots, operational status, and availability of specific services. The dataset also contains some columns with missing values, which were addressed during the data cleaning process.

## Methodology
### 1.Data Cleaning and Preprocessing:

-->Dropped columns with zero non-null values.

-->Imputed missing values for numerical columns using the median.

-->Imputed missing values for categorical columns using the mode.

-->Verified data consistency and completeness.

### 2.Exploratory Data Analysis (EDA):

-->Analyzed key statistics and distributions of numerical and categorical variables.

-->Visualized the geographic distribution of healthcare facilities across provinces and counties.

-->Examined the distribution of facility ownership and operational status.

-->Created visualizations to compare key metrics such as the number of beds and facility types.

### 3.Data Visualization:

-->Used Python libraries (Matplotlib, Seaborn) to create visualizations.

-->Generated histograms, bar plots, and count plots to effectively communicate findings.

## Significance
The analysis of healthcare facility data in Kenya provides valuable insights into the current state of the healthcare sector. The findings can help healthcare administrators make informed decisions to improve healthcare services.

## > Identify any missing values

```python
# To get the columns that have missing values
data.isnull()
```

```python
# Drop the columns with zero non-null values
columns_to_drop = ['ANC', 'BEOC', 'BLOOD', 'CAES SEC', 'CEOC', 'EPI', 'GROWM', 'HCT', 'OPD', 
    'OUTREACH', 'PMTCT', 'RAD/XRAY', 'RHTC/RHDC', 'TB DIAG', 'TB LABS', 'TB TREAT', 'YOUTH'
]
data.drop(columns=columns_to_drop, inplace=True)
```

```python
numerical_columns = ['Beds', 'Cots', 'Official Alternate No']
for column in numerical_columns:
    data[column].fillna(data[column].median(), inplace=True)

categorical_columns = ['Division', 'Location', 'Sub Location', 'Description of Location', 'Nearest Town', 
    'Official Landline', 'Official Fax', 'Official Mobile', 'Official Email', 
    'Official Address', 'Town', 'Post Code', 'In Charge', 'Job Title of in Charge', 
    'Open 24 Hours', 'Open Weekends', 'ART', 'C-IMCI', 'FP', 'HBC', 'IPD'
]
for column in categorical_columns:
    data[column].fillna(data[column].mode(0), inplace=True)
```

```python
# Impute categorical columns with mode
data['Division'].fillna(data['Division'].mode()[0], inplace=True)
data['Location'].fillna(data['Location'].mode()[0], inplace=True)
data['Sub Location'].fillna(data['Sub Location'].mode()[0], inplace=True)
data['Description of Location'].fillna(data['Description of Location'].mode()[0], inplace=True)
data['Nearest Town'].fillna(data['Nearest Town'].mode()[0], inplace=True)
data['Official Landline'].fillna(data['Official Landline'].mode()[0], inplace=True)
data['Official Fax'].fillna(data['Official Fax'].mode()[0], inplace=True)
data['Official Mobile'].fillna(data['Official Mobile'].mode()[0], inplace=True)
data['Official Email'].fillna(data['Official Email'].mode()[0], inplace=True)
data['Official Address'].fillna(data['Official Address'].mode()[0], inplace=True)
data['Town'].fillna(data['Town'].mode()[0], inplace=True)
data['Post Code'].fillna(data['Post Code'].mode()[0], inplace=True)
data['In Charge'].fillna(data['In Charge'].mode()[0], inplace=True)
data['Job Title of in Charge'].fillna(data['Job Title of in Charge'].mode()[0], inplace=True)
data['Open 24 Hours'].fillna(data['Open 24 Hours'].mode()[0], inplace=True)
data['Open Weekends'].fillna(data['Open Weekends'].mode()[0], inplace=True)
data['ART'].fillna(data['ART'].mode()[0], inplace=True)
data['C-IMCI'].fillna(data['C-IMCI'].mode()[0], inplace=True)
data['FP'].fillna(data['FP'].mode()[0], inplace=True)
data['HBC'].fillna(data['HBC'].mode()[0], inplace=True)
data['IPD'].fillna(data['IPD'].mode()[0], inplace=True)
```

```python
# Confirm if the missing values were dropped
data.shape
```

```python
# Get the summary of the dataset after cleaning the dataset
data.info()
```

## > Identify outliers

```python
# Select numerical columns
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Create a box plot for all numerical columns
plt.figure(figsize=(15, 10))
data[numerical_columns].boxplot()
plt.title('Box Plot for All Numerical Columns')
plt.xticks(rotation=45)
plt.show()
```

NOTE:
***Numerical columns were checked for outliers due to having extreme values as compared to categorical columns which do not have extreme values.***

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1_beds = data['Beds'].quantile(0.25)
Q3_beds = data['Beds'].quantile(0.75)
IQR_beds = Q3_beds - Q1_beds

# Define lower and upper bounds for outliers
lower_bound_beds = Q1_beds - 1.5 * IQR_beds
upper_bound_beds = Q3_beds + 1.5 * IQR_beds

# Remove outliers
data = data[(data['Beds'] >= lower_bound_beds) & (data['Beds'] <= upper_bound_beds)]
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1_cots = data['Cots'].quantile(0.25)
Q3_cots = data['Cots'].quantile(0.75)
IQR_cots = Q3_cots - Q1_cots

# Define lower and upper bounds for outliers
lower_bound_cots = Q1_cots - 1.5 * IQR_cots
upper_bound_cots = Q3_cots + 1.5 * IQR_cots

# Remove outliers
data = data[(data['Cots'] >= lower_bound_cots) & (data['Cots'] <= upper_bound_cots)]
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1_alternate_no = data['Official Alternate No'].quantile(0.25)
Q3_alternate_no = data['Official Alternate No'].quantile(0.75)
IQR_alternate_no = Q3_alternate_no - Q1_alternate_no

# Define lower and upper bounds for outliers
lower_bound_alternate_no = Q1_alternate_no - 1.5 * IQR_alternate_no
upper_bound_alternate_no = Q3_alternate_no + 1.5 * IQR_alternate_no

# Remove outliers
data = data[(data['Official Alternate No'] >= lower_bound_alternate_no) & (data['Official Alternate No'] <= upper_bound_alternate_no)]
```

```python
# To confirm if the outliers were removed
data.shape
```

## 1. Get the distribution of beds

```python
# Histogram for Beds
plt.figure(figsize=(10, 6))
sns.histplot(data['Beds'], bins=30, kde=True)
plt.title('Distribution of Beds')
plt.xlabel('Number of Beds')
plt.ylabel('Frequency')
plt.show()
```

## 2. Distribution of Cots

```python
# Histogram for Cots
plt.figure(figsize=(10, 6))
sns.histplot(data['Cots'], bins=30, kde=True)
plt.title('Distribution of Cots')
plt.xlabel('Number of Cots')
plt.ylabel('Frequency')
plt.show()
```

## 3. The Facility Types

```python
# Bar plot for Facility Types
plt.figure(figsize=(12, 8))
sns.countplot(y='Type', data=data, order=data['Type'].value_counts().index)
plt.title('Distribution of Facility Types')
plt.xlabel('Count')
plt.ylabel('Facility Type')
plt.show()
```

## 3. Geographical Distribution of Facilities

```python
# Bar plot for the number of facilities per province
plt.figure(figsize=(12, 8))
sns.countplot(y='Province', data=data, order=data['Province'].value_counts().index)
plt.title('Number of Facilities per Province')
plt.xlabel('Count')
plt.ylabel('Province')
plt.show()

# Bar plot for the number of facilities per county
plt.figure(figsize=(15, 10))
sns.countplot(y='County', data=data, order=data['County'].value_counts().index)
plt.title('Number of Facilities per County')
plt.xlabel('Count')
plt.ylabel('County')
plt.show()
```

## Explanation
**The number of facilities varies significantly across different provinces and counties.
Some provinces and counties have a higher concentration of facilities, while others have fewer facilities.**

## 4. Distribution of Facility based on their Ownership

```python
# Bar plot for facility ownership
plt.figure(figsize=(12, 8))
sns.countplot(y='Owner', data=data, order=data['Owner'].value_counts().index)
plt.title('Distribution of Facility Ownership')
plt.xlabel('Count')
plt.ylabel('Owner')
plt.show()
```

## 5. Operational Status

```python
# Bar plot for operational status
plt.figure(figsize=(10, 6))
sns.countplot(x='Operational Status', data=data)
plt.title('Operational Status of Facilities')
plt.xlabel('Operational Status')
plt.ylabel('Count')
plt.show()
```
